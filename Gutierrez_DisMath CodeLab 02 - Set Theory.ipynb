## Coding Activity: Set Theory

```python
# Favorite Artists in the Music Industry
pop_artists = ["Ariana Grande","Taylor Swift","Katy Perry"]
kpop_artists = ["BTS","Stray Kids","Blackpink","TWICE"]

# Favorite Anime
shoujo_anime = ["Akatsuki no Yona","Ouran High School Club","Toradora!","Fruits Basket"]
shonen_anime = ["Haikyu!!","Jujutsu Kaisen","HunterxHunter","My Hero Academia"]
```

```python
pop = set(pop_artists)
kpop = set(kpop_artists)

shoujo = set(shoujo_anime)
shonen = set(shonen_anime)
```

## Inner Join

```python
# Artists that are both considered as KPOP and Pop
common = kpop.intersection(pop)
common
```

```python
# Anime that are both considered as Shoujo and Shonen
common = shoujo.intersection(shonen)
common
```

## Left Join

```python
# Arists that are under the KPOP Genre
kpop_ljoin = common.union(kpop)
kpop_ljoin
```

```python
# Anime that are under Shoujo Genre
shoujo_ljoin = common.union(shoujo)
shoujo_ljoin
```

## Right Join

```python
# Artists that are under the Pop Genre
pop_ljoin = common.union(pop)
pop_ljoin
```

```python
# Anime that are under Shonen Genre
shonen_ljoin = common.union(shonen)
shonen_ljoin
```

## Left Exclusive Join

```python
# Artists that are KPOP genre but not Pop genre
kpop_exLjoin = kpop.difference(pop)
kpop_exLjoin
```

```python
# Anime that are mostly watched by girls
shoujo_exLjoin = shoujo.difference(shonen)
shoujo_exLjoin
```

## Right Exclusive Join

```python
# Artists that are Pop genre but not KPOP genre
pop_exRjoin = pop.difference(kpop)
pop_exRjoin
```

```python
# Anime that are mostly watched by boys
shonen_exRjoin = shonen.difference(shoujo)
shonen_exRjoin
```

## Full Outer Join

```python
# A list of my favorite artists
popnkpop_FoutJoin = pop.union(kpop)
popnkpop_FoutJoin
```

```python
# A list of my favorite anime
shoujonshonen_FoutJoin = shoujo.union(shonen)
shoujonshonen_FoutJoin
```
